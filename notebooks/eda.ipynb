```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
file_path = '../docs/logs/metrics/training_metrics.csv'
df = pd.read_csv(file_path)

# Set the style of the visualizations
sns.set(style="whitegrid")
```

```python
# Create a figure with multiple subplots
fig, axes = plt.subplots(3, 2, figsize=(15, 15))
fig.suptitle('Training Metrics Over Epochs', fontsize=20)

# Plot average loss
sns.lineplot(x='epoch', y='average_loss', data=df, ax=axes[0, 0], marker='o', color='blue')
axes[0, 0].set_title('Average Loss')
axes[0, 0].set_xlabel('Epoch')
axes[0, 0].set_ylabel('Average Loss')

# Plot average jaccard
sns.lineplot(x='epoch', y='average_jaccard', data=df, ax=axes[0, 1], marker='o', color='green')
axes[0, 1].set_title('Average Jaccard Index')
axes[0, 1].set_xlabel('Epoch')
axes[0, 1].set_ylabel('Average Jaccard Index')

# Plot average dice
sns.lineplot(x='epoch', y='average_dice', data=df, ax=axes[1, 0], marker='o', color='red')
axes[1, 0].set_title('Average Dice Coefficient')
axes[1, 0].set_xlabel('Epoch')
axes[1, 0].set_ylabel('Average Dice Coefficient')

# Plot average pixel accuracy
sns.lineplot(x='epoch', y='average_pixel_accuracy', data=df, ax=axes[1, 1], marker='o', color='purple')
axes[1, 1].set_title('Average Pixel Accuracy')
axes[1, 1].set_xlabel('Epoch')
axes[1, 1].set_ylabel('Average Pixel Accuracy')

# Plot average precision
sns.lineplot(x='epoch', y='average_precision', data=df, ax=axes[2, 0], marker='o', color='orange')
axes[2, 0].set_title('Average Precision')
axes[2, 0].set_xlabel('Epoch')
axes[2, 0].set_ylabel('Average Precision')

# Plot average recall
sns.lineplot(x='epoch', y='average_recall', data=df, ax=axes[2, 1], marker='o', color='brown')
axes[2, 1].set_title('Average Recall')
axes[2, 1].set_xlabel('Epoch')
axes[2, 1].set_ylabel('Average Recall')

# Adjust layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# Plot average validation loss
plt.figure(figsize=(10, 6))
plt.plot(df['epoch'], df['average_validation_loss'], marker='o', color='orange', label='Average Validation Loss')
plt.title('Average Validation Loss Over Epochs', fontsize=16)
plt.xlabel('Epoch', fontsize=14)
plt.ylabel('Average Validation Loss', fontsize=14)

plt.legend()
plt.show()
```

```python
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# Compute the correlation matrix
correlation_matrix = df.corr()

# Set up figure
plt.figure(figsize=(10, 8))

# Draw heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Training Metrics')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Plot average loss
ax.plot(df['epoch'], df['average_loss'], marker='o', color='blue', label='Average Training Loss')

# Plot average validation loss
ax.plot(df['epoch'], df['average_validation_loss'], marker='o', color='orange', label='Average Validation Loss')
ax.set_title('Average Training and Validation Loss Over Epochs', fontsize=16)
ax.set_xlabel('Epoch', fontsize=14)
ax.set_ylabel('Average Loss', fontsize=14)

ax.legend()
plt.show()
```
