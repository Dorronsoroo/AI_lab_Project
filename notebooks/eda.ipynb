```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# CSV file path
csv_file_path = "../docs/logs/metrics/training_loss_steps.csv"

```

```python
# Check if the file exists
if not os.path.isfile(csv_file_path):
    raise FileNotFoundError(f"CSV file '{csv_file_path}' not found.")

data = pd.read_csv(csv_file_path)

epoch = data['epoch']
step = data['step']
loss = data['loss']

# Plot the loss
plt.figure(figsize=(10, 5))
plt.plot(loss, label='Training Loss') #, marker='o', linestyle='-'
plt.xlabel('Steps')
plt.ylabel('Loss')
plt.title('Training Loss over Steps')
plt.legend()
plt.grid(True)
plt.show()
```
